**------------BAR PLOT**

```python
ax = sns.barplot( x = "Month_ID", y = "Revenue", data = raw_data, ci = None, hue = 'Promo')
#MEAN ne SUM !!!
```

```python
ax = sns.barplot( x = "Revenue", y = "Year", data = raw_data, ci = None, hue = 'Promo', orient = 'horizontal')
```


**------------HISTOGRAM**


```python
x = raw_data['Revenue'].array

ax = sns.distplot(x, color = 'darkblue', )
```


**------------BOXPLOT**

```python
x = raw_data['Revenue'].values

sns.boxplot(x)

```

```python

ax = sns.boxplot(x = 'Day_Name', y = 'Revenue', data = raw_data, color = '#57E5D7')

ax = sns.swarmplot(x = 'Day_Name', y = 'Revenue', data = raw_data,color = 'purple')
```

**------------ScatterPlot**

```python
ax = sns.scatterplot(x = 'Marketing Spend', y = 'Revenue', data = raw_data, hue = 'Promo', style ='Promo')
```

```python
ax = sns.scatterplot(x = 'Marketing Spend', y = 'Visitors', data = raw_data, hue = 'Promo', style ='Promo', size = 'Revenue',
                    sizes = (20,300))
```
